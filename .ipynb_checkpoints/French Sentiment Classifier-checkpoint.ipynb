## Data Preprocessing

### In this part, we will perform certain techniques to prepare our data for the next steps. We will define functions for each step to make the process easier

### 1 : Change all the text to lower case

```python
# I know, we have a predefined function, but I just want to make all the process identical
def Upper(text):
  text=text.lower()
  return text
```

### 2- Word Tokenisation :  Basically, tokenisation is the process of breaking a stream of text up into words , for example : "This is a text" => 'This' , 'is' , 'a' ,'text'

```python
def Tokenisation(text):
  tokenized_text=wt(text)
  return tokenized_text
```

```python
Tokenisation("comment t'allez vous ?")
```

#### As you can see, the tokenizer is not working properly for french text, luckely, we can build a better one using "RegexpTokenizer"

```python
from nltk import RegexpTokenizer
toknizer = RegexpTokenizer(r'''\w'|\w+|[^\w\s]''')
```

```python
def FrTokenisation(text):
  tokenized_text=toknizer.tokenize(text)
  return tokenized_text
```

```python
#Here, we import french stop words and add buissness related items
def getStopWordList():
    #read the stopwords file and build a list
    stopWords = []
    stopWords = get_stop_words('french')
    BusinessWords=['vynd',"service",'serveur',"cest","est","a",'ca',
                   "bar","restaurant","jai","oui","non",'c','happy',
                   'hour','happyhour','fast','food','fastfood','plat']
    return BusinessWords + stopWords
```

```python
stopWord=getStopWordList()
def StopWordsAlphaText(tokenized_text):
  filtred_text=[]
  for word in tokenized_text:
    #strip punctuation
    word = word.strip('\'"?,.')
    #check if the word starts with an alphabet
    val = re.search(r"^[a-zA-Z][a-zA-Z0-9]*$", word)
    #ignore if it is a stop word or val is none
    if(word not in stopWord and val is not None):
      filtred_text.append(word)
  return filtred_text
```

### 4- Word Lemmatization : it's the process during wich we reduce the inflected words properly ensuring that the root word belongs to the language. And since we are dealing with a french text, we will use a French lemmatizer linked below  : [GitHub French Lemmatizer](https://github.com/ClaudeCoulombe/FrenchLefffLemmatizer) 

```python
def FrenchLemmatizer (text):
  final_text=[]
  lemmatizer = FrenchLefffLemmatizer()
  for word in text:
    if(word.isalpha()):
        word_Final = lemmatizer.lemmatize(word,'v')
        final_text.append(word_Final)
  return final_text
```

```python
from nltk.stem.snowball import FrenchStemmer
stemmer = FrenchStemmer()
def FrenchStem (text):
    final_text=[]
    lemmatizer = FrenchLefffLemmatizer()
    for word in text:
        if(word.isalpha()):
            word_Final = stemmer.stem(word)
            final_text.append(word_Final)
    return final_text
```

### Everything seems to be working perfectly, so now and for the last step in this section, we need to make sure that our final output is a string format Now it's time to put all the functions we made to work, and apply them to our  dataframe.
```python
for index,row in data.iterrows():
  string=""
  Upper_text = Upper(row['Content'])
  Token_text = Tokenisation(Upper_text)
  Filtered_text = StopWordsAlphaText(Token_text)
  Lemmatized_text = FrenchLemmatizer(Filtered_text)
  for item in Lemmatized_text:
        string=string + " " + item
  print(len(data)-index)
  data.loc[index,"Final_text"]=string
```

```python
from io import StringIO
data = data[pd.notnull(data['word'])]
data['Sentiment_id'] = data['polarity'].factorize()[0]
category_id_df = data[['polarity', 'Sentiment_id']].drop_duplicates().sort_values('Sentiment_id')
category_to_id = dict(category_id_df.values)
id_to_category = dict(category_id_df[['Sentiment_id', 'polarity']].values)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 3))
features = tfidf.fit_transform(data.word).toarray()
labels = data.Sentiment_id
features.shape
```

```python
from sklearn.feature_extraction.text import BaseEstimator
```

```python
from sklearn.feature_selection import chi2
import numpy as np
N = 2
for Product, category_id in sorted(category_to_id.items()):
    features_chi2 = chi2(features, labels == category_id)
    indices = np.argsort(features_chi2[0])
    feature_names = np.array(tfidf.get_feature_names())[indices]
    unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
    bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
    trigrams = [v for v in feature_names if len(v.split(' ')) == 3]
    print("# '{}':".format(Product))
    print("  . Most correlated unigrams:\n. {}".format('\n. '.join(unigrams[-N:])))
    print("  . Most correlated bigrams:\n. {}".format('\n. '.join(bigrams[-N:])))
    print("  . Most correlated trigrams:\n. {}".format('\n. '.join(trigrams[-N:])))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import cross_val_score
```
```python
models = [
    RandomForestClassifier(n_estimators=200, max_depth=10, random_state=0),
    LinearSVC(),
    MultinomialNB(),
    LogisticRegression(random_state=0),
]
CV = 5
cv_df = pd.DataFrame(index=range(CV * len(models)))
entries = []
for model in models:
    model_name = model.__class__.__name__
    accuracies = cross_val_score(model, features, labels, scoring='accuracy', cv=CV)
    for fold_idx, accuracy in enumerate(accuracies):
        entries.append((model_name, fold_idx, accuracy))
cv_df = pd.DataFrame(entries, columns=['model_name', 'fold_idx', 'accuracy'])
import seaborn as sns
sns.boxplot(x='model_name', y='accuracy', data=cv_df)
sns.stripplot(x='model_name', y='accuracy', data=cv_df, 
              size=8, jitter=True, edgecolor="gray", linewidth=2)
plt.show()
```
