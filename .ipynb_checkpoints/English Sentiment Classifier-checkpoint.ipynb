## Data Preprocessing

### 1 : Change all the text to lower case

```python
# I know, we have a predefined function, but I just want to make all the process identical
def Upper(text):
    text=text.lower()
    return text
```

### 2- Word Tokenisation :  Basically, tokenisation is the process of breaking a stream of text up into words , for example : "This is a text" => 'This' , 'is' , 'a' ,'text'

```python
def Tokenisation(text):
    tokenized_text=wt(text)
    return tokenized_text
```

### 3- Remove Stop Words and Non Alpha Text : Nltk provides as with some stop words so here we define a function that returns as a list with those words and othersfrom our business understanding


```python
from nltk.corpus import stopwords
def getStopWordList():
    #read the stopwords file and build a list
    stopWords = ["Vynd",'place','service','food','restaurant','coffee','pizza','time','bar','people','thing','one','price','shop','menu','lol']
    for word in stopwords.words('english'):
        stopWords.append(word)
    return stopWords
```

```python
from nltk.corpus import stopwords
stopwords=stopwords.words('english')
```

```python
stopWord=getStopWordList()
def StopWordsAlphaText(tokenized_text):
    filtred_text=[]
    for word in tokenized_text:
    #strip punctuation
        word = word.strip('\'"?,.')
    #check if the word starts with an alphabet
        val = re.search(r"^[a-zA-Z][a-zA-Z0-9]*$", word)
    #ignore if it is a stop word or val is none
    if(word not in stopWord and val is not None):
        filtred_text.append(word)
    return filtred_text
```

### 4- Word Lemmatization : it's the process during wich we reduce the inflected words properly ensuring that the root word belongs to the language. 
### A little explanation here : we use somethong called "pos_tagger", this tool is used to identify the nature of the word given. We do this process to provide the word's nature to the lemmatizer to obtain the best result possible

```python
def Lemmetizer(text):
    final_text=[]
    tag_map = defaultdict(lambda : wn.NOUN)
    tag_map['J'] = wn.ADJ
    tag_map['V'] = wn.VERB
    tag_map['R'] = wn.ADV
    word_Lemmatized = WordNetLemmatizer()
    for word, tag in pos_tag(text):
        if(word.isalpha()):
            word_Final = word_Lemmatized.lemmatize(word,tag_map[tag[0]])
            final_text.append(word_Final)
    return final_text
```

### Now lets put our functions to the test

```python
test = "THIS is an english Comment ! I am happy that I can write IT and THEN Process IT ! Can you DO THat ???"
```

```python
Upper_test = Upper(test)
print(Upper_test)
```

```python
Token_test = Tokenisation(Upper_test)
print(Token_test)
```

```python
Filtered_test = StopWordsAlphaText(Token_test)
print(Filtered_test)
```

```python
Final_text = Lemmetizer(Filtered_test)
print(Final_text)
```

### Everything seems to be working perfectly, so now and for the last step in this section, we need to make sure that our final output is a string format Now it's time to put all the functions we made to work, and apply them to our  dataframe. We have to make sure that the final output is in a string format

```python
for index,row in data.iterrows():
    string=""
    Upper_text = Upper(row['Content'])
    Token_text = Tokenisation(Upper_text)
    Filtered_text = StopWordsAlphaText(Token_text)
    Lemmatized_text = Lemmetizer(Filtered_text)
    for item in Lemmatized_text:
        string=string + " " + item
    data.loc[index,"Final_text"]=string
```

### reduce the weightage of more common words like (the, is, an etc.) which occur in all document. This is called as TF-IDF i.e Term Frequency times inverse document frequency.
### Wiki : TF_IDF is a numerical statistic that is intended to reflect how important a word is to a document in a collection or corpus

```python
from io import StringIO
data = data[pd.notnull(data['Content'])]
data['Sentiment_id'] = data['Sentiment'].factorize()[0]
category_id_df = data[['Sentiment', 'Sentiment_id']].drop_duplicates().sort_values('Sentiment_id')
category_to_id = dict(category_id_df.values)
id_to_category = dict(category_id_df[['Sentiment_id', 'Sentiment']].values)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 3), stop_words=stopWord)
features = tfidf.fit_transform(data.Content).toarray()
labels = data.Sentiment_id
features.shape
```

```python
from sklearn.feature_selection import chi2
N = 3
for Product, category_id in sorted(category_to_id.items()):
    features_chi2 = chi2(features, labels == category_id)
    indices = np.argsort(features_chi2[0])
    feature_names = np.array(tfidf.get_feature_names())[indices]
    unigrams = [v for v in feature_names if len(v.split(' ')) == 1]
    bigrams = [v for v in feature_names if len(v.split(' ')) == 2]
    trigrams = [v for v in feature_names if len(v.split(' ')) == 3]
    print("# '{}':".format(Product))
    print("  . Most correlated unigrams:\n. {}".format('\n. '.join(unigrams[-N:])))
    print("  . Most correlated bigrams:\n. {}".format('\n. '.join(bigrams[-N:])))
    print("  . Most correlated trigrams:\n. {}".format('\n. '.join(trigrams[-N:])))
```
